```python
# CSV (Comma Seperated Value)
import csv
data = [['name', 'email_id', 'phone_no'],
        ['shlok', 'shlok@gmail.com', 65427543547],
        ['harsh', 'harsh@gmail.com', 75846862863]
] # LIST INSIDE LIST
```

```python
with open('data.csv', 'w', newline='') as f: #We are opening this file in write mode
    writer = csv.writer(f)  # Writer object

    for i in data:
        writer.writerow(i)  #It will read the data line by line
# Also when we will see the folder, we will see that the fole is created
```

```python
with open('data.csv', 'r') as f:
    read_csv = csv.reader(f)

    for i in read_csv:
        print(i)
```
