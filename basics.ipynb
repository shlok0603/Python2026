```python
def send_email(to, subject, body, cc=None, bcc=None, priority="Normal"):
    print(f"Sending email:")
    print(f"To: {to}")
    print(f"Subject: {subject}")
    print(f"Body: {body}")
    print(f"CC: {cc}")
    print(f"BCC: {bcc}")
    print(f"Priority: {priority}")
    print("Email sent successfully!\n")

# ✅ Using positional arguments
send_email("alice@example.com", "Meeting", "Please attend the meeting at 5 PM.")

# ✅ Using keyword arguments (clear, flexible order)
send_email(
    subject="Reminder",
    to="bob@example.com",
    body="Don't forget to submit the report.",
    priority="High",
    cc="manager@example.com"
)

# ✅ Mixing positional and keyword (positional first)
send_email("carol@example.com", "Greetings", "Happy Birthday!", priority="Low")
```

```python
def order_pizza(customer, size="Medium", crust="Regular", toppings=None, extra_cheese=False):
    if toppings is None:
        toppings = []
    print(f"Order Summary for {customer}:")
    print(f"- Size: {size}")
    print(f"- Crust: {crust}")
    print(f"- Toppings: {', '.join(toppings) if toppings else 'No extra toppings'}")
    print(f"- Extra Cheese: {'Yes' if extra_cheese else 'No'}")
    print("Your pizza will be ready soon!\n")

# ✅ Using only defaults + positional
order_pizza("Alice")

# ✅ Using keyword arguments for clarity
order_pizza(
    customer="Bob",
    size="Large",
    crust="Thin",
    toppings=["Pepperoni", "Olives"],
    extra_cheese=True
)

# ✅ Mix positional and keyword
order_pizza("Charlie", toppings=["Mushrooms", "Paneer"])
```

```python
#Getters/Setters
class Student:
    def __init__(self, name, age):
        self.__age = age   # private storage

    @property
    def age(self):   # getter
        return self.__age

    @age.setter
    def age(self, value):  # setter
        if value < 0:
            raise ValueError("Age cannot be negative")
        self.__age = value

s = Student("Shlok", 21)
print(s.age)   # ✅ uses getter
s.age = 25     # ✅ uses setter
# s.age = -5   # ❌ ValueError
```

```python
class User:
    def __init__(self, username, email, password):   # __init__ is the constructor which runs automatically when new object of class is created
        self.username = username                     # saves the username value in the object
        self._email = email                        #  _ is used to make data attribute protected 
        self.password = password
        
    def get_email(self):                    #get_email is an instance method (requires self). It returns the current internal _email value exactly as stored (no cleaning happens here).
        return self._email

    def set_email(self, new_email):          #set_email sets/replaces the internal _email with whatever new_email you provide.
        self._email=new_email
        
user1 = User("Shlok", "Shlok.nandwana@gmail.com     ", 123)      #user1 and user2 are objects of user class
user2 = User("Sandhya", "Sandhya_deora@gmail.com", 456)

#print(user1.get_email())

#user1._email="123456"
user1.set_email("majornandwana131@gmail.com")
print(user1.get_email())
```

```python
from datetime import datetime

class User:
    def __init__(self, username, email, password):
        self.username = username
        self._email = None
        self.password = password
        self.email = email   # this will call the setter at creation

    @property
    def email(self):
        print(f"Email accessed at {datetime.now()}")
        return self._email

    @email.setter
    def email(self, new_email):
        cleaned = new_email.strip().lower()
        if "@" not in cleaned:
            raise ValueError("Invalid email")
        self._email = cleaned


# Usage
user1 = User("Shlok", "Shlok.nandwana@gmail.com     ", 123)
user1.email = " NEW@Example.COM "       # calls the setter
print(user1.email)                      # calls the getter, prints log + value
```
