```python
class bank_account:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance = self.__balance + amount

    def withdraw(self, amount):
        if self.__balance>=amount:
            self.__balance = self.__balance - amount
            return True
        else:
            return False

    def get_balance(self):
        return self.__balance

sandhya = bank_account(100000)
sandhya.get_balance()
```

```python
# (2) Multilevel Inheritance
class class1:
    def class_1(self):
        return 'Hello Class 1'

class class2(class1):
    def class_2(self):
        return 'Hello Class 2'

class class3(class2):
    def class_3(self):
        pass

x = class3()
x.class_1()
```

```python
# (3) Multiple Inheritance
class c1():
    def c_1(self):
        return 'Hello c1'

class c2():
    def c_2(self):
        return 'Hello c2'

class c3(c1, c2):
    pass
x = c3()
print(x.c_1())
print(x.c_2())
```

```python
#------------ABSTRACTION-------------------- It is used
import abc   #abc is a module inside Python Library which contains tools for abstraction      #abc = Abstract Base Classes   . It creates Abstract Based classes and defines Abstract methods

class udp(abc.ABC):      # Here ABC is a class which is used as base class for abstract classes
# ABC comes from module abc  and udp class inherits from ABC so here it treats udp as an abstract class
# This class is meant to be inherited, not used directly.
    
    @abc.abstractmethod
    def students(self):
        pass
        
    @abc.abstractmethod
    def assign(self):
        pass
        
    @abc.abstractmethod
    def marks(self):
        pass

class stu_detail(udp):
    def students(self):
        return 'This is student detail'

    def assign(self):
        return 'This is assignment'

    def marks(self):
        return 'This is Student marks'

class online_session(udp):
    def students(self):
        return 'Online session'

    def assign(self):
        return 'This is Online assignment'
    
    def marks(self):
        return 'This is Online marks'

onl = online_session()
print(onl.assign())

sa = stu_detail()
print(sa.assign())

#Here it means that we have used same name for all method in every class but when we make variable of that class and call that funtion then the data of that function of that particular class will only be printed.
# We have to implement all the methods in the child classes that have been created in the parent class whether you want to call it or not as these are some conditions of the abstract classes.
```

```python
#           ---------------------DECORATOR FUNCTION---------------------------------

#Here I have to decorate my function as i decorate my home my mentioning the start of my function and end of my function and i need to do it repetitively so i can fix this problem
#Created a new function deco(func) and it will take some function as input and inside it create one more function inner_dec() and we will enter the line that i have to continuosly enter and between that i will call the arguement as a function  and at end i will return this inner_dec function
def deco(func):
    def inner_doc():
        print("This is my start of my function")
        func()
        print("This is my end of my function")
    return inner_doc
```

```python
# --------------DECORATORS---------------------------
@deco
def test():
    print('This is my function to test')
    print(4+5)
test()

#Here it is applying the decorator function as @deco and then in deco function when it executes then it goes inside inner_doc executes initial statement then executes the function/logic that we have to use then executes the end statement and then prints every statements
```

```python
#   ------------DECORATOR FUNCTION------------------------------
import time 

def timer_test(func):
    def inner_timer():
        start = time.time()
        print("Function starts at : ", start)
        func()
        end = time.time()
        print("Function end at : ", end)
        print("Total execution time : ", end - start)
    return inner_timer
```

```python
@timer_test
def new():
    print(1+1)
new()
```

```python
@timer_test
def Test():
    for i in range(100000):
        pass
Test()
```

```python
#-----------------STATIC METHOD---------------------------
#I have to create a function and want to access it without creating it's object
#We can do it by using "static method decorator"
class learn1:
    def student_details(self, name, email):
        print(name, email)

    @staticmethod
    def mentor_class(list):
        print(list)

learn1.mentor_class("Sudhanshu")

#This method is used to create utilities like Database connection, File Communication, Database Storage.Here if i have a mentor function and if it is same for all the students then i do not have to create objects for every student individually. So this static method increases reusability
```

```python
#Accessing STATIC METHOD inside CLASS METHOD-----AND-----STATIC inside STATIC-------
#-----------------------STATIC METHOD inside INSTANCE METHOD

class learn2:
    def student_details(self, name, email):
        print(name, email)

    @staticmethod
    def mentor_mail_id(mentor_mail_id):
        print(mentor_mail_id)

    @staticmethod
    def mentor_class(list):
        learn2.mentor_mail_id("sudh@gmail.com")
        print(list)

    @classmethod
    def class_name(cls):
        cls.mentor_class("Sudhanshu")

    #Instance Method   
    def mentor(self, mentor_list):
        print(mentor_list)
        self.mentor_class("Shlok")

learn2.class_name()
learn2.mentor_class('Sudh')

hello = learn2()
hello.mentor("Shlok")
```
